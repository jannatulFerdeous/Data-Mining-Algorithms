```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
from matplotlib import pyplot as plt
from sklearn import tree
import seaborn as sns
```

```python
dt = pd.read_csv("/kaggle/input/heart-disease-cleveland-uci/heart_cleveland_upload.csv")
print(dt.head())
```

```python
x = dt.iloc[:,[0,1,2,3]].values
y = dt.iloc[:,-1].values

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)

```

```python
DT = DecisionTreeClassifier(criterion='entropy',max_depth=3)
DT.fit(x_train, y_train)
Y_pred=DT.predict(x_test)
```

```python

print("Accuracy:", accuracy_score(y_test, Y_pred))
cm=np.array(confusion_matrix(y_test, Y_pred))

sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)

fig, ax = plt.subplots(figsize=(15, 15)) 
tree.plot_tree(DT,ax=ax,feature_names=['sepal length','sepal width','petal length','petal width'])
plt.show()

```

```python
Dt = DecisionTreeClassifier(criterion='entropy',min_samples_split=10)
Dt.fit(x_train, y_train)
Y_pred=Dt.predict(x_test)


print("Accuracy:", accuracy_score(y_test, Y_pred))
cm=np.array(confusion_matrix(y_test, Y_pred))

fig, ax = plt.subplots(figsize=(15, 15)) 
tree.plot_tree(Dt,ax=ax,feature_names=['sepal length','sepal width','petal length','petal width'])
plt.show()
```
